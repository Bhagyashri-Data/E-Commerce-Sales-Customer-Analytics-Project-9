```python
import pandas as pd
import numpy as np
```

```python
sales_data = {
   'Transaction_id' : [101,102,103,104,105,106,107,108,109,110],
    'Cust_id' : [501,502,503,504,505,506,507,508,509,510],
    'Product_category' : ['Electronics','Clothing','Electronics','Home','Clothing','Home','Electronics','Books','Clothing','Books'],
    'Quantity_ordered' : [2,1,np.nan,3,1,2,2,5,1,np.nan],
    'Price_per_unit_INR' : [15000,1200,25000,4500,1200,3000,15000,600,30000,80000]
    }
```

```python
customer_data = {
     'Cust_id' : [501,502,503,504,505,506,507,508,509,510],
    'Customer_Name' : ['Bhagyashri', 'Amit', 'Rahul', 'Priya', 'Sanjay', 'Deepak', 'Neha','Shiv','Ragini','Viraj'],
    'City': ['Pune', 'Mumbai', 'Pune', 'Delhi', 'Mumbai', 'Bangalore', 'Chennai','Karnataka','Delhi','Surat']
}
```

```python
df_sales = pd.DataFrame(sales_data)
df_customer = pd.DataFrame(customer_data)
```

```python
print("Sales DataFrame Loaded:")
print(df_sales.head(5))
print("\ncustomer DataFrame Loaded:")
print(df_customer.head(5))
```

```python
# 1. Checking for missing (NaN) values in the Sales DataFrame
```

```python
print(df_sales.isnull().sum())
```

```python
# 2. Filling the missing values in 'Quantity_ordered' with a default value of 1
# SQL Equivalent: COALESCE(Quantity_ordered, 1)
```

```python
df_sales['Quantity_ordered'] = df_sales['Quantity_ordered'].fillna(1)
```

```python
# 3. Checking the data after filling missing values
print(df_sales.isnull().sum())
```

```python
# 4. Finding and removing duplicate transactions (If any)
# SQL Equivalent: SELECT DISTINCT *
```

```python
(len(df_sales))
```

```python
df_sales= df_sales.drop_duplicates()
```

```python
(len(df_sales))
```

```python
# 1. Renaming column names to make them look cleaner
# SQL Equivalent: SELECT Cust_id AS Customer_ID, Price_per_unit_INR AS Price_INR
```

```python
df_sales = df_sales.rename(columns={
    'Cust_id': 'Customer_id',
    'Price_per_unit_INR' : 'Price_INR'})
```

```python
df_sales
```

```python
df_sales['Quantity_ordered'] = df_sales['Quantity_ordered'].astype(int)
```

```python
(df_sales.dtypes)
```

```python
# Using loc for Multiple Conditions (SQL: WHERE Product_category = 'Electronics' AND Price_INR > 10000)
```

```python
advanced_filter = df_sales.loc[(df_sales['Product_category'] == 'Electronics') & (df_sales['Price_INR'] > 10000)]
```

```python
advanced_filter
```

```python
# SQL Equivalent: SELECT Product_category, SUM(Quantity_ordered), SUM(Price_INR) FROM df_sales GROUP BY Product_category
```

```python
grouping = df_sales.groupby('Product_category').agg (Total_Quantity = ('Quantity_ordered','sum'), Total_Revenue= ('Price_INR','sum')).reset_index()
```

```python
grouping
```

```python
# SQL Equivalent: SELECT * FROM df_sales s INNER JOIN df_customer c ON s.Customer_id = c.Cust_id
# Note: df_customer contains 'Cust_id', so we use left_on and right_on
```

```python
merge_on = pd.merge(df_sales,df_customer, left_on= 'Customer_id', right_on= 'Cust_id',how = 'inner')
```

```python
merge_on
```
